```python
# Conectando ao banco de dados

import pyodbc

conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=localhost;"
    "DATABASE=helpdesk_db;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()
```

```python
from faker import Faker
import random
from datetime import datetime, timedelta
import random

# Dizendo a biblioteca para que use apenas dados brasileiros
fake = Faker("pt_BR")
```

```python
import random

sistema_operacional = ["Windows 07", "Windows 10", "Windows 11"]
fabricante = ["Dell", "Lenovo", "Acer", "Samsung"]
tipo = ["Desktop", "Notebook"]

computadores = []


for i in range(1, 125):
    pc = f"PC-{i:03d}"
    computadores.append(pc)

# Iterando diretamente pelos itens da lista
for pc in computadores:
    sistema = random.choice(sistema_operacional)
    fab = random.choice(fabricante)
    tp = random.choice(tipo)

    cursor.execute("""
        INSERT INTO computadores_tb
        (nome_computador, sistema_operacional, fabricante, tipo)
        VALUES (?, ?, ?, ?)
    """, (pc, sistema, fab, tp))

conn.commit()
```
